```python
import pandas as pd

# 載入 Excel 檔案
excel_file_path = 'C:/Users/wei/Documents/實習/RebasCPBL/中職2020年-逐球數據-20230822.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name1 = 'AWAY-打席資訊'  # 請替換成你要載入的工作表名稱
sheet_name2 = 'HOME-打席資訊' 
```

```python
#計算所有出局總數
def calculate_so_go_count(series):
    so_count = (series == 'SO').sum()
    go_count = (series == 'GO').sum()
    gidp_count = (series == 'GIDP').sum()
    fo_count = (series == 'FO').sum()
    sh_count = (series == 'SH').sum()
    sf_count = (series == 'SF').sum()
    return pd.Series({'SO': so_count, 'GO': go_count ,'GIDP':gidp_count, 'FO':fo_count, 'SH':sh_count, 'SF':sf_count})
```

```python
baseball_cols = ['打席結果','守位投','守位捕','守位一', '守位二','守位三', '守位遊', '守位左', '守位中','守位右'] 
```

## 客隊

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name1)
away
```

```python
away_play = away[baseball_cols]
```

```python
away_play
```

```python
away_play = pd.DataFrame(away_play)
grouped_away_play = away_play.groupby('守位遊')['打席結果'].apply(calculate_so_go_count).reset_index()
grouped_away_play = grouped_away_play.rename(columns={'level_1': '打席結果', '打席結果': '次數'})
grouped_away_play
```

```python
grouped_away_play['次數'] = grouped_away_play['次數'].astype(int)

away_table = pd.pivot_table(grouped_away_play, index='守位遊', columns='打席結果', values='次數', aggfunc='sum', fill_value=0)

# Calculate the final result, which is the sum of GO, SO, and GIDP*2
away_table['總出局數'] = away_table['GO'] + away_table['SO'] + away_table['GIDP']*2 +away_table['SH'] + away_table['SF'] + away_table['FO'] 

# Output the result
away_table
```

## 主隊

```python
home = pd.read_excel(excel_file_path, sheet_name=sheet_name2)
home
```

```python
home_play = home[baseball_cols]
home_play
```

```python
home_play = pd.DataFrame(home_play)

grouped_home_play = home_play.groupby('守位遊')['打席結果'].apply(calculate_so_go_count).reset_index()
grouped_home_play = grouped_home_play.rename(columns={'level_1': '打席結果', '打席結果': '次數'})
grouped_home_play
```

```python
grouped_home_play['次數'] = grouped_home_play['次數'].astype(int)

home_table = pd.pivot_table(grouped_home_play, index='守位遊', columns='打席結果', values='次數', aggfunc='sum', fill_value=0)

# Calculate the final result, which is the sum of GO, SO, and GIDP*2
home_table['總出局數'] = home_table['GO'] + home_table['SO'] + home_table['GIDP']*2 +home_table['SH'] + home_table['SF'] + home_table['FO'] 

# Output the result
home_table
```

```python
# Assuming you have two DataFrames: away_table and home_table

# Merge the two DataFrames based on the '守位遊' column
merged_table = away_table.merge(home_table, on='守位遊', suffixes=('_away', '_home'))

# Sum the columns for 'FO', 'GIDP', 'GO', 'SF', 'SH', 'SO', and '總出局數'
merged_table['FO_total'] = merged_table['FO_away'] + merged_table['FO_home']
merged_table['GIDP_total'] = merged_table['GIDP_away'] + merged_table['GIDP_home']
merged_table['GO_total'] = merged_table['GO_away'] + merged_table['GO_home']
merged_table['SF_total'] = merged_table['SF_away'] + merged_table['SF_home']
merged_table['SH_total'] = merged_table['SH_away'] + merged_table['SH_home']
merged_table['SO_total'] = merged_table['SO_away'] + merged_table['SO_home']
merged_table['總出局數_total'] = merged_table['總出局數_away'] + merged_table['總出局數_home']

# Display the merged and summed table
merged_table

# Get the columns without "_away" and "_home" suffixes
columns_to_keep = [col for col in merged_table.columns if not (col.endswith('_away') or col.endswith('_home'))]

# Create a new DataFrame with only the desired columns
final_table = merged_table[columns_to_keep]

# Display the final table
final_table
```
