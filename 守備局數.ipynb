```python
import pandas as pd
```

```python
excel_file_path = 'C:/Users/ASUS/Documents/實習/Rebas/CPBL_2023.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name1 ='AWAY-打席資訊'  # 請替換成你要載入的工作表名稱
sheet_name2 ='HOME-打席資訊'
```

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name1)
home = pd.read_excel(excel_file_path, sheet_name=sheet_name2)
```

```python
cols = ['開始出局數','結束出局數','守位投','守位捕','守位一', '守位二','守位三', '守位遊', '守位左', '守位中','守位右'] 
```

```python
away_game = away[cols]
home_game = home[cols]
```

```python
away_game
```

```python
summary = pd.concat([home_game,away_game], ignore_index=False)
```

```python
summary['出局數']=summary['結束出局數']-summary['開始出局數']
```

```python
summary 
```

```python
df= summary.groupby(['守位右']).agg({'出局數':'sum'})
df['IP']=df['出局數']/3
df
```

```python
df.to_excel('RF守備局數.xlsx', index=True)
```
